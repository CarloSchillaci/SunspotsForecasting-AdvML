Organize your dataset such that you have time series of a length you can

```python
# we organize the dataset into overlapping time series sequences of a fixed length.
# each sequence will contain 120 months (10 years) of data, based on the periodicity observed in Task 1.

time_series_length = 120

# split the dataset into train and test sets first 
# (we use .iloc[] to preserve the temporal order of the data
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size], df.iloc[train_size:]

# fitting the scaler on the training set
# (not on the test data to avoid data leakage)
scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train.values)
test_scaled = scaler.transform(test.values)
```

```python
# conversion back to DataFrame for consistency
train_scaled = pd.DataFrame(train_scaled, index=train.index, columns=train.columns)
test_scaled = pd.DataFrame(test_scaled, index=test.index, columns=test.columns)
```

```python
def create_time_series(data, time_series_length):
    """
    Converts a 1D time series into overlapping sequences of fixed length.

    Returns:
        X : Input sequences of shape (samples, time_series_length).
        y : Target values of shape (samples,).
        
    """
    
    X, y = [], []
    
    # Loop through the data to create sequences
    for i in range(len(data) - time_series_length):
        
        # Append the input sequence (X) and the corresponding target value (y)
        X.append(data[i:i + time_series_length])
        y.append(data[i + time_series_length])
        
    return np.array(X), np.array(y)
```

```python
X_train, y_train = create_time_series(train_scaled.values, time_series_length)
X_test, y_test = create_time_series(test_scaled.values, time_series_length)

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

### 4.1. Prepare 6-Step Data

```python
future_steps = 6
```

```python
def create_time_series_multi(data, time_series_length, future_steps):
    data = np.array(data)
    X, y = [], []
    for i in range(len(data) - time_series_length - future_steps + 1):
        # Input sequence: [t, ..., t+11]
        X.append(data[i:(i + time_series_length)])
        # Output sequence: [t+12, ..., t+17]
        y_seq = data[i + time_series_length : i + time_series_length + future_steps]
        if y_seq is not None and len(y_seq) == future_steps:  # Ensure valid slice
            y.append(y_seq.flatten())
    # Reshape X to 3D for RNNs
    return np.array(X).reshape(-1, time_series_length, 1), np.array(y)

```

```python
# Generate multi-step data for train and test sets

X_train_multi, y_train_multi = create_time_series_multi(train_scaled, time_series_length, future_steps)
X_test_multi, y_test_multi = create_time_series_multi(test_scaled, time_series_length, future_steps)
```

### 4.2. Approach 1: Iterative Prediction

The iterative prediction approach uses a single-step model to predict one step ahead, then updates the input sequence with the predicted value to forecast subsequent steps. While this method is simple and leverages the trained single-step model, it can accumulate errors over multiple steps. Each prediction depends on the accuracy of the previous step, making the model prone to compounding errors, especially for longer horizons. This approach is computationally efficient but less robust for multi-step forecasting.

```python
def iterative_prediction(model, X_test, steps):
    """
    multi-step predictions using a trained model.
    """
    preds = []
    for i in range(len(X_test)):
        current_seq = X_test[i:i+1]  # Shape: (1, time_series_length, 1)
        future_preds = []
        for _ in range(steps):
            pred = model.predict(current_seq, verbose=0)  # Predict next step
            future_preds.append(pred[0, 0])  # Append the prediction
            # Update the sequence with the predicted value
            pred_reshaped = pred.reshape(1, 1, 1)
            current_seq = np.append(current_seq[:, 1:, :], pred_reshaped, axis=1)
        preds.append(future_preds)
    return np.array(preds)
```

#### 4.2.1: SimpleRNN Iterative Multi-Step

```python
# Train the SimpleRNN model for iterative prediction
iterative_rnn_model = Sequential([
    SimpleRNN(64, activation='relu', return_sequences=False, input_shape=(time_series_length, 1)),
    Dropout(0.2),  # Dropout layer to prevent overfitting
    Dense(1)  
])

iterative_rnn_model.compile(optimizer=tf.keras.optimizers.Adam(1e-3), loss="mse")

history_iterative_rnn = iterative_rnn_model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=1
)

# Perform iterative predictions
iterative_preds_rnn = iterative_prediction(iterative_rnn_model, X_test_multi, steps=future_steps)
```

#### 4.2.2: Dense Iterative Multi-Step

```python
iterative_dense_model = Sequential([
    tf.keras.layers.Flatten(input_shape=(time_series_length, 1)),  # Flatten the input
    Dense(64, activation='relu'), 
    Dropout(0.2),

    Dense(1)  # Output layer for single-step prediction
])
iterative_dense_model.compile(optimizer=tf.keras.optimizers.Adam(1e-3), loss="mse")

history_iterative_dense = iterative_dense_model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=1
)

# Perform iterative predictions
iterative_preds_dense = iterative_prediction(iterative_dense_model, X_test_multi, steps=future_steps)
```

### 4.3. Approach 2: Direct Prediction 


The direct prediction approach trains a model to predict all future steps simultaneously. This method avoids error accumulation by generating predictions for multiple steps in one forward pass. It should be more robust for multi-step forecasting as it does not rely on previous predictions. However, it requires a more complex model architecture and additional training data to learn the relationships between input sequences and multi-step outputs effectively. While computationally more demanding, it often yields better results for longer prediction horizons.

```python
def create_seq2seq_data(data, time_series_length, future_steps):
    """
    Creates data for a seq-to-seq model.
    X shape: (samples, time_series_length, 1)
    y shape: (samples, time_series_length, future_steps)
    """
    data = np.array(data)
    X_list, y_list = [], []
    
    # Stop 'future_steps' early to ensure we can create all targets
    for i in range(len(data) - time_series_length - future_steps + 1):
        
        # 1. Get the input sequence (X)
        X_seq = data[i : i + time_series_length]
        X_list.append(X_seq)
        
        # 2. Get the target sequence (y)
        y_seq_at_each_step = []
        for j in range(time_series_length):
            # For step 'j' in X, the target is [j+1, ..., j+future_steps]
            target_start = i + j + 1
            target_end = i + j + 1 + future_steps
            y_seq_at_each_step.append(data[target_start : target_end].flatten())
        
        y_list.append(np.array(y_seq_at_each_step))

    return np.array(X_list), np.array(y_list)

# Create the new datasets
# Note: These will have slightly fewer samples than before
X_train_s2s, y_train_s2s = create_seq2seq_data(train_scaled.values, time_series_length, future_steps)
X_test_s2s, y_test_s2s = create_seq2seq_data(test_scaled.values, time_series_length, future_steps)

print("X_train_s2s shape:", X_train_s2s.shape)
print("y_train_s2s shape:", y_train_s2s.shape)
print("X_test_s2s shape:", X_test_s2s.shape)
print("y_test_s2s shape:", y_test_s2s.shape)
```

#### 4.3.1. SimpleRNN Direct Multi-Step

```python
# Build the Seq2Seq RNN model
seq2seq_rnn_model = Sequential([
    SimpleRNN(64, activation='relu', return_sequences=True, input_shape=(time_series_length, 1)),
    Dropout(0.2),
    TimeDistributed(Dense(future_steps))  # Output layer
])

seq2seq_rnn_model.compile(optimizer=tf.keras.optimizers.Adam(1e-3), loss="mse")
seq2seq_rnn_model.summary()

# Train the model

history_s2s_rnn = seq2seq_rnn_model.fit(
    X_train_s2s, y_train_s2s,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=1
)

# Predict directly
y_pred_s2s_rnn_all_steps = seq2seq_rnn_model.predict(X_test_s2s)
```

we train the model to forecast the next 10 values at every time step, instead of all that once . This increases the number of error gradients flowing through the model, stabilizing and speeding up training.

#### 4.3.2. Dense Direct Multi-Step

```python
# Reshape y_train_multi to match the model's output shape
y_train_multi_expanded = np.repeat(y_train_multi[:, np.newaxis, :], time_series_length, axis=1)

# Build a TimeDistributed Dense model
seq2seq_dense_model = Sequential([
    TimeDistributed(Dense(64, activation='relu'), input_shape=(time_series_length, 1)),
    TimeDistributed(Dense(future_steps))  # Predicts future_steps at each time step
])

seq2seq_dense_model.compile(optimizer=tf.keras.optimizers.Adam(1e-3), loss="mse")

# Train the model with corrected input and output shapes
history_s2s_dense = seq2seq_dense_model.fit(
    X_train_multi,  # Keep the 3D shape (samples, time_series_length, 1)
    y_train_multi_expanded,  # Reshaped to (samples, time_series_length, future_steps)
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=1
)

# Predict directly
y_pred_s2s_dense_all_steps = seq2seq_dense_model.predict(X_test_multi)
```

```python
def persistence_multi_step(X_test, future_steps):
    """
    Multi-step persistence model: predicts the last observed value in the input sequence
    for all future steps.
    """
    preds = []
    for i in range(len(X_test)):
        # Get the last value of the input sequence
        last_value = X_test[i, -1, 0]
        # Predict the same value for all future steps
        preds.append([last_value] * future_steps)
    return np.array(preds)
```

```python
# Generate predictions using the persistence model
persistence_preds = persistence_multi_step(X_test_multi, future_steps)
```

The persistence is only for comparison

```python
def inverse_transform_multi(data_multi, scaler):
    """Helper function to inverse transform 2D multi-step data."""
    # Get original shape
    n_samples, n_steps = data_multi.shape
    # Reshape to 2D (n_samples * n_steps, 1 feature)
    data_flat = data_multi.reshape(-1, 1)
    # Inverse transform
    data_flat_orig = scaler.inverse_transform(data_flat)
    # Reshape back to (n_samples, n_steps)
    return data_flat_orig.reshape(n_samples, n_steps)
```
